```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test, y_train, y_test = train_test_split(X,y, test_size=.3, random_state=10)
```

```python
cat = [var for var in X_train.columns if df[var].dtype == 'O']
numerical = [var for var in X_train.columns if df[var].dtype != 'O']
```

```python
# handling missing of numerical data
for x in [X_train, X_test]:
    for cal in numerical:
        x[cal].fillna(X_train[cal].mean(), inplace=True)
```

```python
X_test[numerical].isna().sum()
```

```python
# handling missing of cat data
for x in [X_train, X_test]:
    for cal in cat:
        x[cal].fillna(X_train[cal].mode()[0], inplace=True)
```

```python
X_test[cat].isna().sum()
```

```python
# till here, we handled all the missing values 
```

```python
# now will work on outliears 

def capping(d, col, max_value):
    return np.where(d[col]> max_value, max_value, d[col])

for d in [X_train, X_test]:
    d["Rainfall"] = capping(d, "Rainfall", 3.2)
    d["Evaporation"] = capping(d, "Evaporation", 21.8)
    d["WindSpeed9am"] = capping(d, "WindSpeed9am", 55) 
    d["WindSpeed3pm"] = capping(d, "WindSpeed3pm", 57)            
```

```python
#Rainfall, Evaporation, WindSpeed9am, WindSpeed3pm

IQR = df['WindSpeed3pm'].quantile(.75) - df['WindSpeed3pm'].quantile(.25)
lower_limit =df['WindSpeed3pm'].quantile(.25) - (3*IQR)
upper_limit =df['WindSpeed3pm'].quantile(.75) + (3*IQR)

print(IQR, lower_limit, upper_limit)
```
